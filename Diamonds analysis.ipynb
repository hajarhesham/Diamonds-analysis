```python
plt.figure(figsize=(10,8))
sb.regplot(data=df,x='carat',y='price')
```

**insights**


*   at ideal => vs2 the most
*   at premium =>s11 the most
*   at good=>s11 the most
*   at very good=>s11 the most
*   Fair =>s12 the most






```python
plt.figure(figsize=(15,8))
sb.countplot(data=df,x='cut',hue='clarity')
plt.show()
```

**the insights**
*   when the carat is higher the price increase 
*   most depth between 55-66






```python
plt.figure(figsize=(20,5))
sb.regplot(data=df,x='carat',y='depth'
,scatter_kws={"s": df['price']/100})
plt.show()
```

**insight**


*   at cut ideal => most dimond between 0-5000 in prices(low price) (ideal is the most common and the lowest price)
*   most of diamond is ideal 
*   least diamond is the fair



```python
g = sb.FacetGrid(data = df, col = 'cut',size=5)
g.map(plt.hist,'price')
plt.show()
```

```python
g = sb.FacetGrid(data = df, col = 'cut', 
                 palette = 'colorblind',size=5)
g.map(plt.scatter, 'carat', 'price')
```

**insights**


*  the most expensive diamonds are the green and orange except in the fair cuts  
*  the less expensive diamonds are blue(E) and orang except in the fair cuts 






```python
plt.figure(figsize=(15,8))
g = sb.barplot(data = df, x = 'cut', y = 'price', hue = 'color')
g.legend(loc = 0, ncol = 3, framealpha = 10, title = 'color')
```

**insights**

*   the ideal is the less weight that's why its the less expensive 


```python
plt.figure(figsize=(15,8))
sb.barplot(data = df, x = 'cut', y = 'carat',palette=['#34a0a4'])
```
